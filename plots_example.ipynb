# Experimenting with PyPlot and Julia for faster plotting

First we will import relevant libraries

```julia
using Plots
using LaTeXStrings
gr()
```

### Plot 1:

```julia
θ = LinRange(0, 2π, 100)
y₁ = sin.(θ);
```

```julia
plot(θ, y₁,
    title = L"sin(\theta)", 
    xlim = (0, 2π),
    ylim = (-1.25, 1.25),
    xlabel = L"\theta",
    ylabel = "Amplitude",
    legend = false
)

```

### Plot 2:

```julia
x = LinRange(1, 5, 100)
y₂ = x.^2 - 3x .- 2

plot(x, y₂, 
    c = "pink",
    title = L"y = x^2",
    xlabel = L"x"*" mm",
    ylabel = L"y"*" mm",
    legend = false
)
```

### Plot 3:

```julia
scatter(θ, y₁,
	alpha = 0.5,
	color = :jet1,
	marker_z = y₁,
	marker = :hexagon,
	markersize = 5,
	legend = false
)
```

### Unitful

```julia
using Unitful
using Unitful: mm, cm, m, inch, MPa, N
using LinearAlgebra

Unitful.preferunits(u"mm, s"...)
```

```julia

F = [1000, 3200, -4000, 4300, -2570, 6630]N 
```

```julia
r = [300mm, 400mm, 10inch, 15cm, 0.188m, 27cm] .|> mm
```

```julia
scatter(ustrip(r), ustrip(F),
    title = "r vs F",
    xlabel = "r [mm]",
    ylabel = "F [mm]",
    legend = false
)
```

```julia
using DataFrames

df = DataFrame(r=r, F=F)
```
